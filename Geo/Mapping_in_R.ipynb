# Read in & Clean Data
For this analysis we will use data from the National Transportation Safety Board (NTSB). "The NTSB aviation accident database contains information from 1962 and later about civil aviation accidents and selected incidents within the United States, its territories and possessions, and in international waters..."

```r
# Read in Data
url = 'http://app.ntsb.gov/aviationquery/Download.ashx?type=csv'
data <- read.csv(url, header=TRUE, sep="|")
```

```r
summary(data)
```

```r
# Extract & Define Date
data$Date <- mdy(data$Event.Date)

# Subset Dataframe to focus on accidents in 2019
data <- data[data$Date >= "1990-01-01",]
```

# Point & Contour Maps

```r
# Add the geom_point() (you recognize this!) to the map
ggmap(terrain) + geom_point(data = data, # Define your dataset
                        aes(x = Longitude, y = Latitude), # Select the latitude and longitude columns
                        size = 1, # Set the point size
                        shape=19, # Select the desired shape
                        color='red', # Set the desired color
                        alpha=0.1) # Set the desired transparency
```

```r
# Wow, that's a lot of points, what about a stat density?
ggmap(terrain) + stat_density2d(data = data, # Define your dataset
                        aes(x = Longitude, 
                            y = Latitude,
                            fill=..level..,
                            alpha=..level..),
                            size=0.01,
                            bins=10,
                            geom="polygon")+ 
                theme(legend.position="none")
```

```r
# Let's create multiple maps by aircraft type
ggmap(terrain) + stat_density2d(data = data, # Define your dataset
                        aes(x = Longitude, 
                            y = Latitude,
                            fill=..level..,
                            alpha=..level..),
                            size=0.01,
                            bins=10,
                            geom="polygon")+
                  theme(legend.position="none")+
                  facet_wrap(~ Aircraft.Category, ncol=4) # Create multiple maps by aircraft category
                  
```

```r
# Subset our data to only fatal accidents
serious_data <- data[!is.na(data$Total.Fatal.Injuries>=1),]

# Creating the map
ggmap(terrain) + geom_point(data = serious_data, # Define your dataset
                        aes(x = Longitude, 
                            y = Latitude,
                            color=as.factor(Amateur.Built)), # Color by ametur built aircraft or not
                        shape=19, # Select the desired shape
                        alpha=0.5)+
                scale_color_manual(values=c("#FFFFFF", "#4598CF", "#F80D0D"))+

                # Adjusting the theme and axes
                theme_minimal()+
                theme(legend.position="bottom",
                      axis.text.y=element_blank(),
                      axis.text.x=element_blank())+
                xlab('')+
                ylab('')+
                ggtitle('Fatal Accidents Since 1990 - Amateur vs Non-amateur Aircraft')+
                labs(color='Amateur Built Aircraft')+
                guides(color=guide_legend(reverse = TRUE))
```

# Choropleth Maps

```r
# Remove white space
serious_data$Location <- str_trim(serious_data$Location)

# Extract State Code
serious_data$state_abrv <- str_sub(serious_data$Location, -2)
```

```r
# Convert State Code to State Name

# Read in CSV with conversion key
state_names <- read.csv('http://worldpopulationreview.com/static/states/abbr-name.csv', header=FALSE)

# Rename columns
names(state_names) <- c("state_abrv", "state_name")
```

```r
# Use Dplyr to get a (rough) accident count by state
accident_count <- serious_data %>% group_by(state_abrv) %>% tally()

# Join with the state names data
data <- merge(state_names, accident_count, by="state_abrv")

# lower case state name
data <- data %>% mutate(state_name = tolower(state_name))
```
