```r
# Pulling JSON data from Github
library(jsonlite)
library(tidyverse)
url = 'https://raw.githubusercontent.com/vega/vega/master/docs/data/crimea.json'
data <- fromJSON(url, flatten=TRUE)

# Transform from wide to long data
data_long <- gather(data, type, value, disease:wounds, factor_key=TRUE)

# Define plot
ggplot(data_long, aes(x=as.factor(date), y=value, fill=type)) + # Define aesthetics that we're going to use

       geom_bar(stat="identity", position="stack") + # Create a stacked bar chart with our observation values

       scale_fill_manual(values = c("blue",  # Define custom colors for each layer of our stacked bar chart
                                    "grey",
                                     "red")) +

       xlab("Month") + # Define label for x axis

       ylab("Number of Deaths") + # Define label for y axis

      theme(axis.text.x = element_text(angle=90)) + # Rotate date labels 90 degrees

      coord_trans(y = "sqrt") # Transform the coordinate plane to have unequal breaks on the y-axis
```

```r
# Pulling JSON data from Github
library(jsonlite)
library(tidyverse)
url = 'https://raw.githubusercontent.com/vega/vega/master/docs/data/crimea.json'
data <- fromJSON(url, flatten=TRUE)

# Transform from wide to long data
data_long <- gather(data, type, value, disease:wounds, factor_key=TRUE)

# Define plot
ggplot(data_long, aes(x=as.factor(date), y=value, fill=type)) + # Define aesthetics that we're going to use

       geom_bar(stat="identity", position="stack") + # Create a stacked bar chart with our observation values

       scale_fill_manual(values = c("blue",  # Define custom colors for each layer of our stacked bar chart
                                    "grey",
                                     "red")) +

       xlab("Month") + # Define label for x axis

       ylab("Number of Deaths") + # Define label for y axis

       theme_minimal() + # Use a minimal theme to remove the grey background

      theme(axis.text.y = element_blank(), # Clear out y axis labels since they aren't necessary with polar coordinates
           axis.title.y = element_blank())  + 

      coord_polar() # Transform visualization to polar coordinates
```

```r
# Define dataset
data <- gapminder
data = data[data$year==2007,]

# Define & print plot
ggplot(data, aes(lifeExp, gdpPercap, color=continent)) + 
   geom_point(size=2)+
   facet_grid(rows=vars(continent))+
   theme_minimal() # Use a minimal theme for your plots
```

```r
# Define dataset
data <- gapminder
data = data[data$year==2007,]

# Define & print plot
ggplot(data, aes(lifeExp, gdpPercap, color=continent)) + 
   geom_point(size=2)+
   facet_grid(rows=vars(continent))+
   theme_bw()+
   labs(x = 'Life Expectancy', # Assign a new x-axis title
        y = 'GDP per Capita', # Assign a new y-axis title
        title='GDP per Capita by Continent', # Assign a new main title
        subtitle='142 countries in 2007', # Assign a new subtitle
        caption='Data provided by Gapminder') # Assign a new caption
```
